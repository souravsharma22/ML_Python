```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
%matplotlib inline
```

```python
from sklearn.datasets import load_diabetes
```

```python
data = load_diabetes()
```

```python
data
```

```python
print(data.DESCR)
```

```python
x = pd.DataFrame(data=data.data,columns=['age','sex','bmi','bp','s1','s2','s3','s4','s5','s6'])
```

```python
y = data['target']
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
sns.heatmap(x_train.corr(),annot=True)
```

```python
from sklearn.tree import DecisionTreeRegressor
model = DecisionTreeRegressor()
model.fit(x_train,y_train)
```

```python
prediction = model.predict(x_test)
from sklearn.metrics import mean_absolute_error,mean_squared_error
print(mean_absolute_error(y_test,prediction))
print(mean_squared_error(y_test,prediction))
```

## hyperparameter tunning

```python
param = {
    'criterion': ['squared_error', 'friedman_mse', 'absolute_error', 'poisson'],
    'splitter': ['best', 'random'],
    'max_depth':[1,2,3,4,5,6,7,10,15,20,25],
    'max_features':['auto','sqrt','log2']
}
```

```python
from sklearn.model_selection import GridSearchCV
tune_model = DecisionTreeRegressor()
grid = GridSearchCV(tune_model,param_grid=param,cv=5,scoring='neg_mean_squared_error')
import warnings
warnings.filterwarnings('ignore')
grid.fit(x_train,y_train)
```

```python
grid.best_params_
```

```python
best_model = grid.best_estimator_
```

```python
prediction1 = grid.predict(x_test)
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
print(mean_absolute_error(y_test,prediction1))
print(mean_squared_error(y_test,prediction1))
print(r2_score(y_test,prediction1))
```

```python
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(best_model,filled=True)
```
