```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
print(iris.DESCR)
```

```python
print(iris['target'])
```

```python
x = pd.DataFrame(iris['data'],columns=["sepal length in cm",
    "sepal width in cm"
    , "petal length in cm"
    , "petal width in cm"])
```

```python
x
```

```python
y = iris['target']
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
model.fit(x_train,y_train)
```

```python
prediction1=model.predict(x_test)
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
print(classification_report(prediction1,y_test))
print(accuracy_score(prediction1,y_test))
print(confusion_matrix(prediction1,y_test))
```

```python
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(model,filled=True)
```

## postprunning the tree

```python
post_prun_model = DecisionTreeClassifier(max_depth=2)
post_prun_model.fit(x_train,y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(post_prun_model,filled=True)
```

```python
prediction=post_prun_model.predict(x_test)
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
print(classification_report(prediction,y_test))
print(accuracy_score(prediction,y_test))
print(confusion_matrix(prediction,y_test))
```
