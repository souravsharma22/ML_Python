```python
import seaborn as sns
import pandas as pd
```

```python
df = sns.load_dataset('tips')
```

```python
df.head()
```

```python
gen_bill = df.groupby('time')['total_bill'].mean().to_dict()
```

```python
df['time_encoded']= df['time'].map(gen_bill)
```
