```python
list_marks = [33,45,32,56,67,78,89,90,87,76,65,54,43,99,88,77]
min,q1,mid,q2,maz= np.quantile(list_marks,[0,0.25,0.50,0.75,1.0])
```

```python
min,q1,mid,q2,maz
```

```python
IQR = q2-q1
lower_fence = q1-1.5*(IQR)
higher_fence= q2+1.5*(IQR)
```

```python
lower_fence,higher_fence
```
