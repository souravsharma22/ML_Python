```python
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
```

```python
df.head()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
encoder = OneHotEncoder()
```

```python
gender_encode = encoder.fit_transform(df[['sex']]).toarray()
```

```python
import pandas as pd
gender_encode_df = pd.DataFrame(gender_encode,columns=encoder.get_feature_names_out())
```

```python
gender_encode_df
```

```python
smoker_encode = encoder.fit_transform(df[['smoker']]).toarray()
smoker_encode_df = pd.DataFrame(smoker_encode,columns=encoder.get_feature_names_out())
```

```python
smoker_encode_df
```

```python
day_encode = encoder.fit_transform(df[['day']]).toarray()
day_encode_df = pd.DataFrame(day_encode,columns=encoder.get_feature_names_out())
```

```python
day_encode_df
```

```python
size_encode = encoder.fit_transform(df[['size']]).toarray()
size_encode_df = pd.DataFrame(size_encode,columns=encoder.get_feature_names_out())
```

```python
size_encode_df
```

```python
final_df= pd.concat([gender_encode_df,smoker_encode_df,day_encode_df,size_encode_df],axis=1)
```
