```python
from sklearn.datasets import make_classification
```

```python
x,y=make_classification(n_samples=1000,n_redundant=0,n_features=2,n_clusters_per_class=1,weights=[0.90],random_state=12,n_informative=2)
```

```python
import pandas as pd
```

```python
df1= pd.DataFrame(x,columns=["feature 1","feature 2"])
df2 = pd.DataFrame(y,columns=["target"])
```

```python
final_df = pd.concat([df1,df2],axis=1)
```

```python
final_df.head()
```

```python
final_df['target'].value_counts()
```

```python
import matplotlib.pyplot as plt
plt.scatter(final_df['feature 1'], final_df['feature 2'], c=final_df['target'])
```

```python
from imblearn.over_sampling import SMOTE
```

```python
oversample = SMOTE()
```

```python
x,y = oversample.fit_resample(final_df[['feature 1', 'feature 2']], final_df['target'])
```
