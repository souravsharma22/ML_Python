```python
import matplotlib.pyplot as plt
```

```python
x = [2,3,4,5,6]
y=[4,9,16,25,36]
plt.plot(x,y)
```

```python
# we can define our x and y lable, this is a line plot
plt.plot(x,y)
plt.xlabel("Numbers")
plt.ylabel("square of the numbers")
plt.title("Basic Line plot")
plt.show()
```

```python
# we can design and customize this plot
plt.plot(x,y,color="red", linestyle="--", marker= 'o', linewidth=3, markersize=7)
plt.grid(True)
plt.xlabel("Numbers")
plt.ylabel("square of the numbers")
plt.title("Basic Line plot")
plt.show()
```

```python
# in plt.subplot(2,4,3) : the 2 (first number ) give us nno of rowa in the plot
# 4(secound no.) tells about the no of column
# 3(third number ) tells about the position of the current plot 
x = [2,3,4,5,6]
y=[4,0,16,0,36]
z =[6,-9,12,-15,18]

plt.figure(figsize=(9,7))

plt.subplot(2,2,1)
plt.plot(x,y,color="red", linestyle="--", marker= 'o', linewidth=1, markersize=7)
plt.title("Plot 1")
plt.grid(True)

plt.subplot(2,2,2)
plt.plot(x,z,color="green", linestyle="--", marker= '*', linewidth=1, markersize=7)
plt.title("Plot 2")
plt.grid(True)

plt.subplot(2,2,3)
plt.plot(y,z,color="green", linestyle="--", marker= '*', linewidth=1, markersize=7)
plt.title("Plot 3")
plt.grid(True)

plt.subplot(2,2,4)
plt.plot(z,y,color="red", linestyle="--", marker= 'o', linewidth=1, markersize=7)
plt.title("Plot 4")
plt.grid(True)

```
