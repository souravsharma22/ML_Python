```python
import pandas as ps
import seaborn as sn
import matplotlib.pyplot as plt
```

# SeaBorn Basic Ploting

```python
# loading data set from library
df = sn.load_dataset("tips")
```

```python
# we can load dataset and use them from seaborn
df.head()
```

```python
sn.scatterplot(x='total_bill', y= "tip",data=df)
plt.title("PLot between total bill vs tip")
```

```python
sn.lineplot(x= "size",y= "total_bill",data=df)
```

```python
sn.barplot(x='day',y='total_bill',data=df)
```

```python
# similarly we have other plots like boxplot
sn.boxplot(x='day',y='total_bill',data=df)
```

```python
sn.violinplot(x='day',y='total_bill',data=df)
```

```python
# all pyplot visiualization we can create using seaborn
sn.histplot(x='day',y='total_bill',data=df) # this is a 2d histplot
```

```python
sn.histplot(df["total_bill"],bins=10,kde=True)
```

```python
# we have also other visualizations 
sn.kdeplot(df["total_bill"],fill=True)
```

```python
sn.pairplot(df)
```

```python
df.isnull().sum()
```

```python
corr1 = df[['total_bill','tip','size']].corr()
sn.heatmap(corr1,cmap="coolwarm")
```
