```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import load_breast_cancer
df = load_breast_cancer()
```

```python
df.keys()
```

```python
print(df.DESCR)
```

```python
cancer_df = pd.DataFrame(df['data'],columns=df['feature_names'])
```

```python
cancer_df.head()
```

```python
#standerisation
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
scaler.fit(cancer_df)
```

```python
scaled_data = scaler.transform(cancer_df)
```

```python
scaled_data
```

```python
#applying pc algorithms for feature extraction
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
```

```python
pca_data=pca.fit_transform(scaled_data)
```

```python
pca_data
```

```python
pca.explained_variance_
```

```python
plt.figure(figsize=(8,6))
plt.scatter(pca_data[:,0],pca_data[:,1], c = df['target'],cmap='plasma')
plt.xlabel("first Priciple component")
plt.ylabel("Second Priciple component")
```
