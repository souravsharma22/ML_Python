# Histogram

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data1 = [1,2,3,3,4,5,6,7,8,9,8,7,6,5,4,2,3]
plt.hist(data1,bins=5, color="red",edgecolor = "black")
```

# pichart

```python
size = [200,309,285,500,115,150]
label = ["A","B","c","d","e","f"]
colors=["gold","yellow","lightgreen","red","lightcoral","yellowgreen"]
explode =(0.2,0,0,0.3,0,0)
plt.pie(size,labels=label,colors=colors,shadow=True,autopct="%1.1f%%",explode=explode)
```

```python
category =[278,350,700,1000,1300,1500,1700]
marking = ["very Low","Low","medium","upper","high","Excelent"]
df["Category"] =pd.cut(df["Proline"],bins=category,labels=marking,right=False)
```

```python
df.head()
```

```python
df2 = df.groupby("Category",observed =True)["Mg"].sum()
print(df2)
```

```python
df2.plot(kind='bar',color="red")
```

```python
df3 = df.groupby("Category",observed =True)["Phenols"].sum().reset_index()
```

```python
df3
```

```python
df3.plot(kind="line",color="blue")
```
