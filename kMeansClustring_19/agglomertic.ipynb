```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
%matplotlib inline
```

```python
iris = load_iris()
```

```python
iris_data= pd.DataFrame(iris['data'])
```

```python
iris_data.head()
```

```python
iris_data.columns = iris.feature_names
```

```python
iris_data.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
x_scaled=scaler.fit_transform(iris_data)
```

```python
from sklearn.decomposition import PCA
```

```python
psa = PCA(n_components=2)
```

```python
psa_scaled=psa.fit_transform(x_scaled)
```

```python
plt.scatter(psa_scaled[:,0], psa_scaled[:,1],c = iris['target'])
```

```python
import scipy.cluster.hierarchy as super
#Making the dendrogram
plt.figure(figsize=(20,7))
plt.title("Dendogram")

#creating dendrogram
super.dendrogram(super.linkage(psa_scaled,method='ward'))
plt.xlabel("sample index")
plt.ylabel("Sample index")
```

```python
from sklearn.cluster import AgglomerativeClustering
cluster = AgglomerativeClustering(n_clusters=2,  linkage='ward')
cluster.fit(x_scaled)
```

```python
cluster.labels_
```

```python
plt.scatter(psa_scaled[:,0], psa_scaled[:,1],c = cluster.labels_)
```
